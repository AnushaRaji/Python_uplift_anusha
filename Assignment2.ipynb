```python
#Exercise 2: 
# Create a dictionary for one department with keys "budget" and "employees"
# "employees" should be a list of two dictionaries, each with "name" and "salary"
# Example:
#     {
#        "Engineering": {
#            "budget": 500000,
#            "employees": [
#                {"name": "Alice", "salary": 90000},
#                {"name": "Bob", "salary": 85000}
#            ]
#        },
#        "Marketing": { ... }
#    }
# Calculate and print the total salary, then print if it is within the budget

department = {
    "Sales " : {
        "budget": 300000,
        "employees": [
            {"name": "John", "salary": 70000},
            {"name": "Jane", "salary": 75000}
        ]
    },
    "HR" : {
        "budget": 200000,
        "employees": [
            {"name": "Doe", "salary": 60000},
            {"name": "Smith", "salary": 65000}
        ]
    },
    "Engineering":{
        "budget":250000,
        "employees":[
            {"name":"Alice","salary":90000},
            {"name":"Bob","salary":85000}
        ]
    }
}
for dept_name, dept_info in department.items():
    budget = dept_info["budget"]
    employees = dept_info["employees"]
    total_salary = sum(emp["salary"] for emp in employees)
    print(f"Department: {dept_name}")
    print(f"Total Salary: ${total_salary}")
    if total_salary <= budget:
        print("Within Budget")
    else:
        print("Over Budget")
    print()  # Blank line for readability
```

```python
# Exercise 3: 
# Create a list of 3 coordinate tuples (x, y)
# Print the first and last coordinates
# Create and print a new list with x and y swapped in each tuple

coordinates = [(2, 3), (5, 7), (8, 1)]
print(f"First coordinate: {coordinates[0]}")
print(f"Last coordinate: {coordinates[-1]}")

swapped_coordinates = [(y, x) for (x, y) in coordinates]
print("Swapped coordinates:", swapped_coordinates)
```

```python
#Exercise 4:
# Create two sets: tech_subscribers and news_subscribers with some overlapping emails
# Print the emails that are in both sets
# Print all unique emails from both sets
# Remove one email from both sets and print the updated sets
tech_subscribers = {"test1 ", "test2" , "test3"}
news_subscribers = {"test2 ", "test4" , "test5"}  
common_emails = tech_subscribers.intersection(news_subscribers)
all_unique_emails = tech_subscribers.union(news_subscribers)
print("Common emails:", common_emails)
print("All unique emails:", all_unique_emails)
```

```python
# Exercise 5: 
# Create a dictionary "library" with 3 books; each value is a dictionary with "title", "author", "available"
# Create a list "borrowing_history" with 2 tuples (book_id, user_name)
# Print all titles of books that are available
# Print all user names that appear in borrowing_history
library = {
    1: {"title": "1984", "author": "George Orwell", "available": True},
    2: {"title": "To Kill a Mockingbird", "author": "Harper Lee", "available": False},
    3: {"title": "The Great Gatsby", "author": "F. Scott Fitzgerald", "available": True}
}
borrowing_history = [(1, "Alice"), (2, "Bob")]
available_books = [book_info["title"] for book_info in library.values() if book_info["available"]]
print("Available books:", available_books)  
user_names = {user_name for (_, user_name) in borrowing_history}
print("User names in borrowing history:", user_names)
```
