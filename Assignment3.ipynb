```python
# Write a function `count_long_lines(filename, limit)` that:
# 1. Opens a text file and reads all lines.
# 2. Uses a **for-loop** to count how many lines have length > limit.
# 3. Uses an **if-statement** inside the loop to check line length.
# 4. Returns the count.
# Then, write code that:
# - Prompts the user for a filename and a limit.
# - Calls the function.
# - Prints the result.

def count_long_lines(filename, limit):
    count = 0
    with open(filename, 'r') as file:
        lines = file.readlines()
        print(lines)  # Print the lines read from the file
        for line in lines:
            if len(line) > limit:
                count += 1
        print(count)
    return count    
# Prompt user for filename and limit
filename = input("Enter the filename: ")
limit = int(input("Enter the length limit: "))
# Call the function and print the result
result = count_long_lines(filename, limit)
print(f"Number of lines longer than {limit}, characters: {result}")
# This script creates a text file and defines a function to count long lines in a file.
```

```python
# Exercise 2:

# Write a script that:
# 1. Opens a file named "numbers.txt" in write mode.
# 2. Uses a **while-loop** to repeatedly ask the user for integers.
# 3. Stops when the user enters "stop".
# 4. Writes each valid integer to the file (one per line).
# After writing:
# - Re-open the file in read mode.
# - Use a for-loop to read each number and compute the sum.
# - Print the total sum.
with open("numbers.txt", "w") as file:
   while True:
      user_input = input("Enter an integer (or 'stop' to finish): ")
      
      if user_input.lower() == 'stop':
         break
      else:
         number = int(user_input)
         file.write(f"{number}\n")   

total_sum = 0
with open("numbers.txt", "r") as file:
    for line in file:
        total_sum += int(line.strip())
        print(line.strip())
print(f"The total sum of the numbers is: {total_sum}")
```

```python
# Exercise 3:

# Create a function `filter_words(input_file, output_file, min_length)` that:
# 1. Opens input_file, reads all lines.
# 2. For each line, splits into words.
# 3. Uses a loop + if-statement to keep only words whose length >= min_length.
# 4. Writes these filtered words into output_file (one line per original line).
# Then call the function using any sample input/output filenames.
def filter_words(input_file, output_file, min_length):
       with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
        for line in infile:
            words = line.split()
            filtered_words = [word for word in words if len(word) >= min_length]
            outfile.write(' '.join(filtered_words) + '\n')
# Call the function with sample filenames and minimum length
filter_words('hello.txt', 'output.txt', 10)
```

```python
# Exercise 4:

# Write code that:
# 1. Opens a file "temperatures.txt" which contains one temperature per line.
# 2. Reads all lines using readlines().
# 3. Converts them to floats.
# 4. Uses a function `average_temp(data)` to compute the average.
# 5. Prints:
# - The average temperature
# - "High" if avg > 25
# - "Moderate" if 15 <= avg <= 25
# - "Low" if avg < 15
with open("temperatures.txt", "w") as file:
    file.write("30.5\n22.0\n15.5\n10.0\n28.3\n18.7\n")
def average_temp(data):
    return sum(data) / len(data) if data else 0
with open("temperatures.txt", "r") as file:
    lines = file.readlines()
    temperatures = [float(line.strip()) for line in lines]
    avg_temp = average_temp(temperatures)
    print(f"Average Temperature: {avg_temp:.2f}°C")
    if avg_temp > 25:
        print("High")
    elif 15 <= avg_temp <= 25:
        print("Moderate")
    else:
        print("Low")
        
```

```python
#Exercise 5:

# Write a program that:
# 1. Opens a file "logs.txt" and reads each line.
# 2. Each line is formatted like: "LEVEL: message"
# 3. Use a function `count_level(filename, level)` that:
# - Opens the file
# - Counts how many lines start with the given level (e.g., "ERROR")
# - Uses loops + if to check matching lines
# 4. Print how many ERROR, WARNING, and INFO lines are in the file.

def count_level(filename, level):
    count = 0
    with open(filename, 'r') as file:
        for line in file:
            if line.startswith(level + ":"):
                count += 1
    return count
# Create a sample logs.txt file
with open("logs.txt", "w") as file:
    file.write("ERROR: Disk full\n")
    file.write("WARNING: High memory usage\n")
    file.write("INFO: System rebooted\n")
    file.write("ERROR: Failed to load module\n")
    file.write("INFO: Update completed\n")
    file.write("WARNING: Low battery\n")    
# Count and print the number of each log level
error_count = count_level("logs.txt", "ERROR")
warning_count = count_level("logs.txt", "WARNING")
info_count = count_level("logs.txt", "INFO")
print(f"ERROR lines: {error_count}")
print(f"WARNING lines: {warning_count}")
print(f"INFO lines: {info_count}")  

```
