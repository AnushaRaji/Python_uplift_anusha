```python
age_str= input("Please enter your age: ")
age_int= int(age_str)
if age_int < 0:
    print("Error: Age cannot be negative.") 
elif age_int < 18:
    print("You are a minor.")   
elif age_int < 65:
    print("You are an adult.")

test=input("Enter a test score (0-100): ")
test=int(test)+10
print("Your adjusted test score is:", test)
```
