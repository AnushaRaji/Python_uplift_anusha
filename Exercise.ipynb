```python
# Exercise 1: Create a simple tip calculator
# Ask for bill amount and tip percentage
# Display the tip amount and total formatted to 2 decimal places

bill_amount = float(input("Enter the bill amount: "))
tip_percentage = float(input("Enter the tip percentage (e.g., 5 for 5%): "))
tip_amount = bill_amount * (tip_percentage / 100)
total_amount = bill_amount + tip_amount 
print(f"(Bill amount : ${bill_amount:.2f})")
print(f"Tip amount: ${tip_amount:.2f}")
print(f"Total amount: ${total_amount:.2f}") 
```

```python
# Exercise 4: Given a list of numbers from 0 to 9, print:
# - The first three numbers
# - The last four numbers
# - Every third number
# - The list reversed
numbers = list(range(10))
print(f"The first three numbers: {numbers[:3]}")          
print(f"The last four numbers : {numbers[-4:]}")        
print(f"Every third number: {numbers[::3]}")        
print(f"List reversed {numbers[::-1]}")       
```
